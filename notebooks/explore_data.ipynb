```python
from pathlib import Path

import polars as pl
```

```python
current_directory = Path.cwd()

if current_directory.name == "notebooks":
    project_root = current_directory.parent
else:
    project_root = current_directory

raw_data_directory = project_root / "data" / "raw"

print("Current directory:", current_directory)
print("Project root:", project_root)
print("Raw data directory:", raw_data_directory)
```

```python
for file_path in raw_data_directory.glob("*.parquet"):
    print(file_path.name)
```

```python
schedules = pl.read_parquet(
    raw_data_directory / "schedules.parquet"
)

team_stats = pl.read_parquet(
    raw_data_directory / "team_stats.parquet"
)

player_stats = pl.read_parquet(
    raw_data_directory / "player_stats.parquet"
)
```

```python
print("Schedules:", schedules.shape)
print("Team stats:", team_stats.shape)
print("Player stats:", player_stats.shape)
```

```python
completed_games = schedules.filter(
    pl.col("home_score").is_not_null()
)

upcoming_games = schedules.filter(
    pl.col("home_score").is_null()
)

print("Completed games:", completed_games.height)
print("Upcoming games:", upcoming_games.height)
```

```python
upcoming_games.select(
    [
        "season",
        "week",
        "gameday",
        "away_team",
        "home_team",
    ]
).head(20)
```

```python
completed_games = schedules.filter(
    pl.col("home_score").is_not_null()
    & pl.col("away_score").is_not_null()
)

completed_games.select(
    [
        "season",
        "week",
        "home_team",
        "away_team",
        "home_score",
        "away_score",
    ]
).head()
```

```python
completed_games = completed_games.with_columns(
    (pl.col("home_score") > pl.col("away_score"))
    .cast(pl.Int8)
    .alias("home_win")
)
```

```python
completed_games.select(
    [
        "home_team",
        "away_team",
        "home_score",
        "away_score",
        "home_win",
    ]
).head(10)
```

```python
completed_games = completed_games.filter(
    pl.col("home_score") != pl.col("away_score")
)
```
